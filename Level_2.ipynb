```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
```

```python
df = pd.read_csv("Desktop\Data Science\Dataset_.csv")
print(df)
```

```python
# Determine the percentage of restaurants that offer table booking and online delivery.

Table_Booking_Percentage = len(df[df["Has Table booking"]=="Yes"])/len(df)*100
Online_Delivery_Percentage = len(df[df["Has Online delivery"]=="Yes"])/len(df)*100

print(f"Percentage of restaurants that offer table booking = {Table_Booking_Percentage:.2f}%")
print(f"Percentage of restaurants that offer online delivery = {Online_Delivery_Percentage:.2f}%")
```

```python
# Compare the average ratings of restaurants with table booking and those without.

# Group by 'Has Table booking' and calculate the mean rating
average_ratings = df.groupby('Has Table booking')['Aggregate rating'].mean()

# Create a bar plot
plt.bar(average_ratings.index, average_ratings.values)
plt.xlabel('Table Booking')
plt.ylabel('Average Rating')
plt.title('Comparison of Average Ratings (With Table Booking Vs Without Table Booking)')
plt.xticks([0, 1], ['Without Table Booking', 'With Table Booking']) 
plt.show()
```

```python
# Analyze the availability of online delivery among restaurants with different price ranges.

online_delivery_by_price_range = df.groupby('Price range')['Has Online delivery'].value_counts(normalize=True).unstack() * 100
print("Percentage of restaurants offering online delivery by price range:")
print(online_delivery_by_price_range)


# Plot the availability of online delivery by price range
online_delivery_by_price_range.plot(kind='bar')
plt.xlabel('Price Range')
plt.ylabel('Number of Restaurants with Online Delivery')
plt.title('Availability of Online Delivery by Price Range')
plt.show()
```

```python
# availability of online delivery among restaurants with different price ranges.
delivery_cross_tab = pd.crosstab(df['Price range'], df['Has Online delivery'])

# Plot for the comparison
plt.figure(figsize=(10, 6))
delivery_cross_tab.plot(kind='bar', stacked=True, color=['skyblue', 'lightcoral'])
plt.title('Availability of Online Delivery Across Different Price Ranges')
plt.xlabel('Price Range')
plt.ylabel('Number of Restaurants')
plt.legend(title='Online Delivery', labels=['No Online Delivery', 'Online Delivery'])
plt.show()
```

```python
# Create a box plot
sns.boxplot(x='Has Online delivery', y='Price range', data=df)
plt.xlabel('Online delivery')
plt.ylabel('Price range')
plt.title('Distribution of Price range by Availability of online delivery')
plt.show()
```

Task 2 : Price Range Analysis

* Determine the most common price range among all the restaurants.
* Calculate the average rating for each price range.
* Identify the color that represents the highest average rating among different price ranges.

```python
# Determining the most common price range among all the restaurants.

most_common_price_range = df['Price range'].mode().iloc[0]
print(f"Most Common Price Range: {most_common_price_range}")
```

```python
# Calculate the average rating for each price range.

avg_ratings = df.groupby('Price range')['Aggregate rating'].mean()
print(avg_ratings)

plt.bar(avg_ratings.index, avg_ratings.values, color='skyblue', edgecolor='black')
plt.xlabel("Numeric Price Range")
plt.ylabel("Average Rating")
plt.title("Average Rating by Numeric Price Range")
plt.xticks(avg_ratings.index)  # Show numeric price range values on the X-axis
plt.show()
```

```python
# Identify the color that represents the highest average rating among different price ranges.


# Calculate average ratings per price range
avg_ratings = df.groupby('Rating color')['Aggregate rating'].mean()
print(avg_ratings)

# Find the highest average rating
highest_avg_rating = avg_ratings.max()
highest_avg_price_range = avg_ratings.idxmax()  

plt.bar(avg_ratings.index, avg_ratings.values, edgecolor='black')
plt.xlabel("Price Range")
plt.ylabel("Average Rating")
plt.title("Average Rating by Price Range")
plt.show()
```

Task 3 : Feature Engineering

* Extract additional features from the existing columns, such as the length of the restaurant name or address.
* Create new features like "Has Table Booking" or "Has Online Delivery" by encoding categorical variables.

```python
# Extract additional features from the existing columns, such as the length of the restaurant name or address.

df['Restaurant name length'] = df['Restaurant Name'].apply(len)
df['Address length'] = df['Address'].apply(len)
df[['Restaurant Name','Address','Restaurant name length','Address length']]
```
