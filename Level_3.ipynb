Task 1 : Predictive Modeling

* Build a regression model to predict the aggregate rating of a restaurant based on available features.
  Split the dataset into training and testing sets and evaluate the model's performance using appropriate metrics.

```python
df.columns
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error,r2_score
from math import sqrt
```

```python
features=['Country Code','Longitude','Latitude','Average Cost for two','Price range','Votes']
target='Aggregate rating'


df=df.dropna(subset=features+[target])
```

```python
X = df[features]
y = df[target]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,random_state=0)
```

```python
#Linear Regression
model = LinearRegression()
model.fit(X_train, y_train)
model_Prediction=model.predict(X_test)
mse = mean_squared_error(y_test, model_Prediction)
    
r2 = r2_score(y_test, model_Prediction)
print(f"Model: {model}")
print(f"Mean Squared Error: {mse:.4f}")
print(f"R-squared score: {r2:.4f}\n")

#Decison Tree
model = DecisionTreeRegressor(random_state=0)
model.fit(X_train, y_train)
model_Prediction=model.predict(X_test)
mse = mean_squared_error(y_test, model_Prediction)

r2 = r2_score(y_test, model_Prediction)
print(f"Model: {model}")
print(f"Mean Squared Error: {mse:.4f}")
print(f"R-squared score: {r2:.4f}\n")

#Random Forest
model = RandomForestRegressor(random_state=0)
model.fit(X_train, y_train)
model_Prediction=model.predict(X_test)
mse = mean_squared_error(y_test, model_Prediction)

r2 = r2_score(y_test, model_Prediction)
print(f"Model: {model}")
print(f"Mean Squared Error: {mse:.4f}")
print(f"R-squared score: {r2:.4f}\n")


```

Based on above result, the Random Forest Regressor performs the best with highest R score 0.9571

Task 2: Customer Preference Analysis

* Analyze the relationship between the type of cuisine and the restaurant's rating.
* Identify the most popular cuisines among customers based on the number of votes.
* Determine if there are any specific cuisines that tend to receive higher ratings.

```python
#Analyze the relationship between the type of cuisine and the restaurant's rating.

# Calculate average ratings by cuisine
avg_ratings_by_cuisine = df.groupby("Cuisines")["Aggregate rating"].mean()
print("Average rating for each cuisine:")
print(avg_ratings_by_cuisine)
```

```python
# Identify the most popular cuisines among customers based on the number of votes.

most_popular_cuisines = df.groupby('Cuisines')['Votes'].sum().sort_values(ascending=False)
most_popular_cuisines
```

```python
#Determine if there are any specific cuisines that tend to receive higher ratings.

High_rating_cusiones=df.groupby("Cuisines")['Aggregate rating'].mean().nlargest(50)
print("\ntop cuisines based on rating:")
print(High_rating_cusiones)
```

```python
#Task 3: Data Visualization
```

```python
#Create visualizations to represent the distribution of ratings using different charts (histogram, bar plot, etc.).

plt.figure(figsize=(20, 5))
plt.hist(df['Aggregate rating'], bins=30,edgecolor='k')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.title('Distribution of Ratings')
plt.show()

# Bar plot for count rating
plt.figure(figsize=(20, 5))
df['Aggregate rating'].value_counts().sort_index().plot(kind='bar')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.title('Count of Each Rating')
plt.show()

```

```python
# Compare the average ratings of different cuisines or cities using appropriate visualizations.

avg_rating_by_cuisines=df.groupby('Cuisines')['Aggregate rating'].mean().sort_values(ascending=False).head(35)

plt.figure(figsize=(20,5))
sns.barplot(x=avg_rating_by_cuisines.index, y=avg_rating_by_cuisines.values, palette='viridis')
plt.xlabel('Cuisine')
plt.ylabel('Average Rating')
plt.title('Average Ratings by Cuisine')
plt.xticks(rotation=90)
plt.show()

plt.figure(figsize=(20,5))
avg_rating_by_city=df.groupby("City")['Aggregate rating'].mean().sort_values(ascending=False).head(35)
sns.barplot(x=avg_rating_by_city.index, y=avg_rating_by_city.values, palette='viridis')
plt.xlabel('City')
plt.ylabel('Average Rating')
plt.title('Average Ratings by City')
plt.xticks(rotation=90)
plt.show()
```
